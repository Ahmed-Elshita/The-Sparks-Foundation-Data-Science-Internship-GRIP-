```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/GRIP/TASK 5/Indian Premier League/matches.csv')
df1.head()
```

```python
df1.info()
```

```python
df1.isna().sum()
```

```python
df1.drop('umpire3',axis=1,inplace=True)
```

```python
df1.dropna(inplace=True)
df1.isna().sum()
```

```python
df1.shape
```

```python
df1['winner'].value_counts().plot(kind='bar')
plt.xticks(rotation = 90)
plt.xlabel('teams',fontsize=22)
plt.ylabel('Counts',fontsize=22)
plt.title('count of times where teams winner',fontsize=30)
plt.show()
```

```python
sns.barplot(df1['toss_winner'].value_counts()[:10].values, df1['toss_winner'].value_counts()[:10].index,palette='rocket')
plt.title('Top 10 Teams Toss_winner',fontsize=30)
plt.xlabel('Count' ,fontsize=20)
plt.ylabel('toss_winner',fontsize=20)
plt.show()
```

```python
df1['player_of_match'].value_counts()[:10].plot(kind='bar')
plt.xticks(rotation = 90)
plt.xlabel('player',fontsize=22)
plt.ylabel('Counts',fontsize=22)
plt.title('top 10 best players',fontsize=30)
plt.show()
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/GRIP/TASK 5/Indian Premier League/deliveries.csv')
df2.head()
```

```python
df2.info()
```

```python
df2.drop(['fielder','dismissal_kind'],axis=1,inplace=True)
df2['player_dismissed'].fillna(value='NA', inplace=True)
```

```python
df2.dropna(inplace=True)
df2.isna().sum()
```

```python
d2 = df2.groupby(['match_id','inning','batsman'])['total_runs'].count()
d2 = pd.DataFrame(d2)
d2
```

```python
delivery=pd.merge(df2, df1, left_on='match_id', right_on='id')
delivery.head()
```

```python
delivery.drop_duplicates()
```

```python
print('Shape:', delivery.shape)
print('Size:', delivery.size)
```

```python
delivery.isnull().sum()
```

```python
plt.figure(figsize=(15,15))
hm = sns.heatmap(delivery.corr(), annot=True, linewidth=1, cmap='twilight_shifted')
```

```python
plt.figure(figsize=(10,10))
sns.countplot(data=delivery, y='venue')
plt.xlabel('count',fontsize=22)
plt.ylabel('venue',fontsize=22)
plt.title('venues count',fontsize=30)
plt.show()
```

```python
delivery.toss_decision.value_counts().plot(kind='bar')
plt.xlabel('toss_decision',fontsize=22)
plt.ylabel('counts',fontsize=22)
plt.title('Deciding Whether to Bat or Field After Winning the Toss',fontsize=30)
plt.show()
```

```python
delivery.groupby('batsman')['batsman_runs'].agg("sum").sort_values(ascending= False).head().plot(kind='bar')
plt.xlabel('Batsmen',fontsize=22)
plt.ylabel('batsman_runs',fontsize=22)
plt.title('Top 5 Batsmen',fontsize=30)
plt.show()
```

```python
delivery.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head(5).plot(kind='bar')
plt.xlabel('bowler',fontsize=22)
plt.ylabel('player_dismissed',fontsize=22)
plt.title('Top 5 Bowlers',fontsize=30)
plt.show()
```
