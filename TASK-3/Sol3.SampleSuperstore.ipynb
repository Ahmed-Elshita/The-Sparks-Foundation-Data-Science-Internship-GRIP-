Exploratory Data Analysis 

**Sample Superstore**

Import and read data

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns 

# Load the iris dataset
data = '/content/drive/MyDrive/Colab Notebooks/GRIP/TASK 3/SampleSuperstore.csv'
df = pd.read_csv(data)
df.head() # See the first 5 rows
```

```python
df.info()
```

```python
df.columns
```

exploratory using pivot table to show relation between region and sum of profit

```python
RegionProfit=df.pivot_table(index='Region',values= 'Profit',aggfunc='sum',fill_value=0)
RegionProfit
```

```python
sns.barplot(x= 'Region',y= 'Profit',data=df,ci=None)
plt.show()
```

exploratory using pivot table to show relation between region and sum of Sales

```python
RegionSales=df.pivot_table(index='Region',values= 'Sales',aggfunc='sum',fill_value=0)
RegionSales
```

```python
sns.barplot(x= 'Region',y=  'Sales',data=df,ci=None)
plt.show()
```

exploratory using pivot table to show relation between sum of profit to Ship Mode for each Segment

```python
profitForSegment=df.pivot_table(index='Ship Mode',columns= 'Segment',values= 'Profit',aggfunc='sum',fill_value=0)
profitForSegment
```

```python
sns.catplot(x='Ship Mode',y= 'Profit',col= 'Segment',data=df,kind='bar',ci=None)
plt.show()
```

exploratory using pivot table to show relation between sum of Sales to Ship Mode for each Segment

```python
SalesForSegment=df.pivot_table(index='Ship Mode',columns= 'Segment',values= 'Sales',aggfunc='sum',fill_value=0)
SalesForSegment
```

```python
sns.catplot(x='Ship Mode',y=  'Sales',col= 'Segment',data=df,kind='bar',ci=None)
plt.show()
```

exploratory using pivot table to show relation between Category and Sales

```python
SalesForCategory=df.pivot_table(index='Category',values= 'Sales',aggfunc='sum',fill_value=0)
SalesForCategory
```

```python
sns.catplot(x= 'Category',y= 'Sales',data=df,kind='bar',ci=None)
plt.show()
```

exploratory using pivot table to show relation between Category and Profit

```python
ProfitForCategory=df.pivot_table(index='Category',values= 'Profit',aggfunc='sum',fill_value=0)
ProfitForCategory
```

```python
sns.catplot(x= 'Category',y=  'Profit',data=df,kind='bar',ci=None)
plt.show()
```
